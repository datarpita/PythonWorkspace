Suppose we don't want to consider those (rare) words which have appeared only in 3% of the documents, or say those (extremely common ones) which have appeared in 80% of the documents.

Use CountVectorizer(stop_words='english', min_df=.03, max_df=.8) to create a new vocabulary from the training set. What is the size of the new vocabulary?

```python
movie_vec1 = CountVectorizer(stop_words='english',min_df=.03, max_df=.8)
movie_vec1.fit(movie_train_df['text'])
movie_vec1.vocabulary_
len(movie_vec1.vocabulary_)
```

```python
#movie_vec1.fit(movie_test_df['text'])
#movie_vec1.vocabulary_
movie_test_transformed = movie_vec1.transform(movie_test_df['text'])
movie_test_transformed

### 51663
```

Train a Bernoulli Naive Bayes model on the training set and predict the classes of the test set. Each movie review in the test set has been labelled as 'Pos' or 'Neg'. What is the accuracy of the model?

```python
X_movie_train = movie_train_df['text']
Y_movie_train = movie_train_df['class'].map({'Neg':0,'Pos':1})
```

```python
Y_movie_train.head()
```

```python
X_movie_test = movie_test_df['text']
Y_movie_test = movie_test_df['class'].map({'Neg':0,'Pos':1})
```

```python
movie_vec1.fit(X_movie_train)
movie_vec1.vocabulary_
```

```python
X_train_transformed = movie_vec1.transform(X_movie_train)
X_test_tranformed =movie_vec1.transform(X_movie_test)
```

```python
# instantiate bernoulli NB object
bnb = BernoulliNB()

# fit 
bnb.fit(X_train_transformed,Y_movie_train)

# predict class
y_pred_class = bnb.predict(X_test_tranformed)

# predict probability
y_pred_proba =bnb.predict_proba(X_test_tranformed)
```

```python
metrics.accuracy_score(Y_movie_test, y_pred_class)
```

How many reviews are actually negative but have been classified as positive by the model?

```python
cf= metrics.confusion_matrix(Y_movie_test, y_pred_class)
```

```python
TN=cf[0,0]
FN=cf[1,0]
FP=cf[0,1]
TP=cf[1,1]
print ('TN=',TN,' FN=',FN,' FP=',FP,' TP=',TP)
```
