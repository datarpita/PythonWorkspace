## Derived Metrics - exercise

```python
odi_batting_df = pd.read_csv('odi-batting.csv')
```

```python
#odi_batting_df['noOfCenturies'] = np.floor(odi_batting_df['Runs']/100)
def checkCentury(runs):
    if runs >= 100 :
        return 1
    else :
        return 0
    
    
odi_batting_df['noOfCenturies']= odi_batting_df['Runs'].apply(checkCentury)
```

```python
odi_batting_df.head(10)
```

```python
#Which player scored the highest number of centuries?

groupbyplayer = odi_batting_df.groupby(by='Player')
#groupbyplayer.groups
#groupbyplayer.get_group('Sachin R Tendulkar')
groupbyplayer['noOfCenturies'].sum().to_frame().sort_values(by='noOfCenturies', ascending=False)
```

```python
xx = odi_batting_df.loc[odi_batting_df['Player']=='Aamir Sohail']
xx
#xx['noOfCenturies'].sum()
```

```python
#The strike rate of a batsman is defined as the (number of runs scored / number of 100 balls faced). 
#It is a measure of how fast a batsman has played.
#For example, if a batsman scores 40 runs in 50 balls, the strike rate is 80 runs/100 balls.

odi_batting_df['strikeRate']= (100*odi_batting_df['Runs'])/odi_batting_df['Balls']
odi_batting_df.head(10)
```

```python
#Among all the innings where a century was scored by a batsman, which one has the highest strike rate?
odi_batting_century_df = odi_batting_df.loc[odi_batting_df['Runs'] >= 100]
```

```python
odi_batting_century_df.sort_values(by='strikeRate', ascending=False)
```

```python
odi_batting_india_df = odi_batting_df.loc[odi_batting_df['Country']=='India']
```

```python
#odi_batting_india_df['MatchYear'] = odi_batting_india_df['MatchDate'].str.split('-').str[2]
#e.g.
#df['date'] = pd.to_datetime(df['date'])
#df['year'], df['month'] = df['date'].dt.year, df['date'].dt.month
odi_batting_india_df['MatchDate'] = pd.to_datetime(odi_batting_india_df['MatchDate'])
odi_batting_india_df['MatchYear'], odi_batting_india_df['MatchMonth'] = odi_batting_india_df['MatchDate'].dt.year, odi_batting_india_df['MatchDate'].dt.month
```

```python
odi_batting_india_df.head(10)
```

```python
#In which year were the maximum number of centuries scored by Indian players?
odi_batting_century_grouped=odi_batting_india_df.groupby(by=['MatchYear'])
odi_batting_century_grouped['noOfCenturies'].sum().to_frame().sort_values(by='noOfCenturies', ascending=False)
```
