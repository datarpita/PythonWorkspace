## SVM Hyperparameter Tuning - Diabetes Data

Find the optimal value of the hyperparameter C using GridSearchCV(), and then build a linear SVM model using the optimal value of C to predict whether a given patient has diabetes.

```python
# tune the model
```

```python
# specify the number of folds for k-fold CV
n_folds = 5
```

```python
# specify range of parameters (C) as a list
params = {"C": [1, 5, 10, 20, 25, 30]}
```

```python
# create SVC object
linear_model = SVC(kernel='linear')
```

```python
# set up grid search scheme
model_cv = GridSearchCV(estimator = linear_model, 
                         param_grid = params, 
                         scoring= 'accuracy', 
                         cv = n_folds, 
                         verbose = 1,
                        return_train_score=True)   
          
```

```python
# fit the model on n_folds
model_cv.fit(x_train, y_train)
```

```python
# store CV results in cv_results
cv_results = model_cv.cv_results_
print(pd.DataFrame(model_cv.cv_results_))
```

```python
# plot C versus train and test scores (code already written)
plt.figure(figsize=(8, 6))
plt.plot(cv_results['param_C'], cv_results['mean_test_score'])
plt.plot(cv_results['param_C'], cv_results['mean_train_score'])
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.legend(['test accuracy', 'train accuracy'], loc='upper left')
plt.xscale('log')
plt.show()
plt.savefig('hyperparam_c.png') 
```

```python
# best hyperparameter C and best (test) accuracy
best_score = model_cv.best_score_
best_C = model_cv.best_params_['C']
print(best_score)
print(best_C)
```

```python
#  choose the best value of C 
C = best_C 
```

```python
# build a model with best value of C
linear_model = SVC(C=best_C)
linear_model.fit(x_train, y_train)
```

```python
# make predictions
print(diabetes_test.head())
predictions = linear_model.predict(diabetes_test.iloc[:, :7])
predictions[:5]
```

```python
# write columns id, predictions into the output file
d = pd.DataFrame({'id': diabetes_test['id'], 'Diabetes_Predicted': predictions})
print("\n", "d", "\n", d.head())
```

```python
# write the output
d.to_csv('diabetes_predictions.csv', sep=",")
```
