```python
## Selecting and filtering the high value customers. Storing in a new dataframe 'telecom_high_val_cust_df'
```

```python
var_70th_perc = ((telecom_copy_df['total_rech_amt_6'] +  telecom_copy_df['total_rech_amt_7'])/2).quantile(0.7)
var_70th_perc

def isHighValueCust(x):
    if (x>var_70th_perc):
        return 'Y'
    else:
        return 'N'


telecom_copy_df['High_Value_Cust'] = telecom_copy_df['total_rech_amt_9'].map(isHighValueCust)
```

```python
telecom_high_val_cust_df = telecom_copy_df.loc[telecom_copy_df['High_Value_Cust']=="Y"]
telecom_high_val_cust_df.shape
```
