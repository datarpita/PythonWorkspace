```python
def unique_value_columns(df):
    column_unique = []
    for column in df.columns:
        if (df[column].nunique() == 1):
            column_unique.append(column)
    return column_unique
```

```python
column_unique = unique_value_columns(telecom_copy_df)
column_unique
```

```python
def isHighValueCust(x,y):
    print (x, '****', y)
    if ((x>=var_70th_perc) or (y>=var_70th_perc)):
        return 'Y'
    else:
        return 'N'
   

#telecom_copy_df['High_Value_Cust'] = telecom_copy_df['total_rech_amt_9'].map(isHighValueCust)
telecom_copy_df['High_Value_Cust'] = np.vectorize(isHighValueCust)(telecom_copy_df['total_rech_amt_6'],telecom_copy_df['total_rech_amt_7'])
```

```python
telecom_high_val_cust_df = telecom_copy_df.loc[telecom_copy_df['High_Value_Cust']=="Y"]
telecom_high_val_cust_df.shape
```

```python
# Tag high value customers churn as 1 or 0
```

```python
def findchurn(w,x,y,z):
    if (((w==0) or (x==0)) and ((y==0) or (z==0))):
        return 1
    else:
        return 0
    
    

telecom_high_val_cust_df['churn'] = np.vectorize(findchurn)(telecom_high_val_cust_df['total_ic_mou_9'],telecom_high_val_cust_df['total_og_mou_9'], telecom_high_val_cust_df['vol_2g_mb_9'], telecom_high_val_cust_df['vol_3g_mb_9'])
```

```python
#Remove attributes corresponding to churn phase i.e. september month
not_september_cols = [col for col in telecom_high_val_cust_df.columns if '_9' not in col]
not_september_cols
telecom_high_val_cust_minus_sep_df = telecom_high_val_cust_df[not_september_cols]
```
