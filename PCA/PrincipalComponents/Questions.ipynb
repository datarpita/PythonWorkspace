```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('Ratings.csv')
df.head()
```

```python
pca = PCA(svd_solver='randomized', random_state=42)
```

```python
pca.fit(df)
```

```python
pca.components_
```

Once you perform PCA on the given dataset and project the data, what is the approximate variance explained by the first principal component?

```python
pca.explained_variance_ratio_
```
