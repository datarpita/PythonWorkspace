```python
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
```

```python
gold.set_index('Country', inplace=True)
silver.set_index('Country', inplace=True)
bronze.set_index('Country', inplace=True)
```

```python
total = gold.add(silver, fill_value=0).add(bronze, fill_value=0)
total.sort_values(by='Medals', ascending=False)
```

```python
df = pd.read_csv('https://query.data.world/s/Hfu_PsEuD1Z_yJHmGaxWTxvkz7W_b0')
print('Original::',df.shape)
print(round(100*(df.isnull().sum(axis=1)/len(df.index)),2))
df = df[df.isnull().sum(axis=1) <= 5]
print('Modified::',df.shape)
print(round(100*(df.isnull().sum()/len(df.index)),2))
```

```python
df = pd.read_csv('https://query.data.world/s/Hfu_PsEuD1Z_yJHmGaxWTxvkz7W_b0')
mean1 = df['Product_Base_Margin'].mean()
print(mean1)
df.loc[np.isnan(df['Product_Base_Margin']),['Product_Base_Margin']]=mean1
print(round(100*(df.isnull().sum()/len(df.index)),2))
```

```python
#Given a single integer n, create an (n x n) 2D array with 1 on the border and 0 on the inside. 
n = int(input())
arr = np.full((n,n),0)
arr[0,:]=1
arr[n-1,:]=1
arr[:,0]=1
arr[:,n-1]=1
print(arr)
    
```
