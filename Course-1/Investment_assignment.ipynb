### Checkpoint 2 - Table 2.1

```python
master_frame.info()
```

```python
# Clean data : Replace NANs with 0 
master_frame['raised_amount_usd'].fillna(0, inplace=True)
```

```python
# Group the combined company details based on the funding type
all_groups = master_frame.groupby(by='funding_round_type')
all_groups
all_groups.groups.keys()
```

```python
#1) Average funding amount of venture type
venture_group = all_groups.get_group('venture')
print ('The representative value of venture type:\n ',venture_group['raised_amount_usd'].agg([np.mean,np.median,np.std]))
```

```python
#2) Average funding amount of angel type
angel_group = all_groups.get_group('angel')
print ('The representative value of angel type:\n ',angel_group['raised_amount_usd'].agg([np.mean,np.median,np.std]))
```

```python
#3) Average funding amount of seed type
seed_group = all_groups.get_group('seed')
print ('The representative value of seed type:\n ',seed_group['raised_amount_usd'].agg([np.mean,np.median,np.std]))
```

```python
#4) Average funding amount of private equity type
pvt_eqt_group = all_groups.get_group('private_equity')
print ('The representative value of private equity type:\n ',pvt_eqt_group['raised_amount_usd'].agg([np.mean,np.median,np.std]))
```

```python
#5) Considering that Spark Funds wants to invest between 5 to 15 million USD per  investment round, which investment type 
# is the most suitable for them?
data = [['venture',venture_group['raised_amount_usd'].agg(np.mean)],
        ['angel', angel_group['raised_amount_usd'].agg(np.mean)],
        ['seed',seed_group['raised_amount_usd'].agg(np.mean)], 
        ['private_equity',pvt_eqt_group['raised_amount_usd'].agg(np.mean)]]
four_groups = pd.DataFrame(data, columns=['funding_round_type','mean_raised_amount_usd'])
four_groups
```

```python
# Which investment type do you think is the most suitable for Spark Funds?
result1= four_groups.loc[(four_groups['mean_raised_amount_usd']>=5000000) | (four_groups['mean_raised_amount_usd']<=15000000)]
print (result1.loc[result1['mean_raised_amount_usd']==result1['mean_raised_amount_usd'].max()])
```

```python
# Read the mapping.csv file into 'sector_mapping_df' dataframe
sector_mapping_df = pd.read_csv('mapping.csv')
sector_mapping_df.replace(np.NAN, 'unknown', inplace=True)
```

```python
sector_mapping_df.set_index('category_list', inplace=True)
```

```python
#Create a new column 'main_sector' that will contain the name of the main sector for each category
#User-defined function to extract the column name where the value is 1 
def getColName(row):
    #print('****',row.name)
    is_data_present = (sector_mapping_df.loc[row.name])==1    
    #print('^^^^',is_data_present.idxmax())
    colName =  is_data_present.idxmax()
    #print('^^^^',colName)
    return colName

sector_mapping_df['main_sector'] = sector_mapping_df.apply(getColName,axis=1)
```

```python
sector_mapping_df.reset_index(inplace=True)
sector_mapping_df.head(10)
```

```python
#A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and 
#the average amount of investment in each funding type. 
#This chart should make it clear that a certain funding type (FT) is best suited for Spark Funds
```

```python
temp = four_groups.loc[(four_groups['funding_round_type']!='angel')] # select FT seed, venture and private_equity
temp1 = temp.groupby(by='funding_round_type')
temp2 = temp1[['mean_raised_amount_usd']].sum().add_prefix('sum_of_')
plot1_df = temp2.apply(lambda x: 100 * x / float(x.sum()))
plot1_df = plot1_df.reset_index()
```

```python
plt.title('Perc of average investment for funding type')
sns.barplot(x='funding_round_type', y='sum_of_mean_raised_amount_usd', data=plot1_df)
fig1 = plt.figure()
plt.show()
```
