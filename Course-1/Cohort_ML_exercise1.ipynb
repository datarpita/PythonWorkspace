```python
import numpy as np
import pandas as pd
```

```python
cohort_ml_df = pd.read_csv('ML+C5+Know+Your+Cohort.csv')
cohort_ml_df.info()
```

```python
cohort_ml_df.describe()
```

```python
#Replace the vaue '10+' in 'YearsOfExp' to '10'
cohort_ml_df.loc[cohort_ml_df['YearsOfExp']=='10+','YearsOfExp']=10
cohort_ml_df
```

```python
#What is the average experience (approx.) of this cohort
cohort_ml_df['YearsOfExp']=cohort_ml_df['YearsOfExp'].astype(int) #Changing the data type of 'YearsOfExp' column from string to int
cohort_ml_df['YearsOfExp'].mean()
```

```python
#How many students of this cohort have more than 10 years of experience (approx.)?
count = cohort_ml_df.loc[cohort_ml_df['YearsOfExp']>=10,'YearsOfExp'].count()
perc = (count / cohort_ml_df['YearsOfExp'].count() ) * 100
print (count)
```

```python
#What is the share of female candidates in this cohort (choose the nearest value) ?
totalStudents = cohort_ml_df['Gender'].count()
totalStudents
femaleStudents = cohort_ml_df.loc[cohort_ml_df['Gender'].notnull() & 
                          ((cohort_ml_df['Gender']=='female') | (cohort_ml_df['Gender']=='Female'))
                          ,'Gender'].count()
femaleStudents
percFemaleStudents = (femaleStudents /totalStudents) * 100
print (percFemaleStudents)
```

```python
#Find the approximate percentage of students from Bangalore
cohort_ml_df.head()
cohort_ml_df['City ']=cohort_ml_df['City '].str.upper()
studentsInBL = cohort_ml_df.loc[(cohort_ml_df['City ']=='BANGALORE') | (cohort_ml_df['City ']=='BENGALURU'),'City '].count()
percStudInBL = (studentsInBL / totalStudents) * 100
print (percStudInBL)
```

```python
#Which city has the second highest number of learners in the cohort?
citygrps= cohort_ml_df.groupby('City ')
print(citygrps.agg(np.size))
```
